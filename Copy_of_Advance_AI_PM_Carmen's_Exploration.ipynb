**Try plotting a scatterplot after generating x amount of data points across 3 dimensions. Play around the with number of data points (i.e. 100’)**

```python
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()
# Map each onto a scatterplot we'll create with Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="AI PM students years of experience")
plt.show()
```

**Play around with the parameters**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='lower right')

# Major grid lines
plt.grid(True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)

# Minor grid lines
plt.minorticks_on()
plt.grid(True, which='minor', color='beige', alpha=0.8, linestyle='-', lw=1)

# Show the plot
plt.show()
```
