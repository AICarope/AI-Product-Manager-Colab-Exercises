```python
#Play around with the parameters
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='lower right')
#major grid lines
plt.grid(visible=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
#minor grid lines
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)
plt.show()
For the homework please upload the link of your colab
```

```python
# Generate data
x = np.linspace(0, 7, 100)

# Plot sine and cosine
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')

# Add legend
plt.legend(handles=[line1, line2], loc='lower right')

# Major grid lines
plt.grid(True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)

# Minor grid lines
plt.minorticks_on()
plt.grid(True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)

# Show plot
plt.show()
```
